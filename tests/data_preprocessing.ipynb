# Cleaning the data all at once

```python
# Index(['uuid', 'sourceCode', 'title', 'description', 'minimumYearsExperience',
#        'shiftPattern', 'schemes', 'skills', 'otherRequirements', 'ssocCode',
#        'occupationId', 'ssocVersion', 'workingHours', 'numberOfVacancies',
#        'psdUrl', 'categories', 'employmentTypes', 'positionLevels', 'status',
#        'ssecEqa', 'ssecFos', 'postedCompany', 'hiringCompany',
#        'screeningQuestions', 'address', 'metadata', 'salary', '_links'],
#       dtype='object')


company_uens = []
company_names= []
company_descriptions = []

salary_maxs = []
salary_mins= []
salary_types = []

for index, row in df_clean.iterrows():
    # break

    # skills
    skill_dict = row['skills']
    skill_list = [item['skill'] for item in skill_dict]
    df_clean.at[index, 'skills'] = skill_list

    # categories
    categories_dict = row['categories']
    categories_list = [item['category'] for item in categories_dict]
    df_clean.at[index, 'categories'] = categories_list

    # employmentTypes
    employment_dict =  row['employmentTypes']
    employment_list = [item['employmentType'] for item in employment_dict]
    df_clean.at[index, 'employmentTypes'] = employment_list

    # positionLevels
    postion_dict = row['positionLevels']
    postion_list = [item['position'] for item in postion_dict]
    df_clean.at[index, 'positionLevels'] = postion_list

    # status
    status_dict = row['status']
    status = status_dict['jobStatus']
    df_clean.at[index, 'status'] = status

    #postedCompany
    company_dict = row['postedCompany']
    company_uen = company_dict['uen']
    company_name = company_dict['name']
    company_description = company_dict['description']
    company_uens.append(company_uen)
    company_names.append(company_name)
    company_descriptions.append(company_description)

    #screeningQuestions
    question_dict = row['screeningQuestions']
    question_list = [item['question'] for item in question_dict]
    df_clean.at[index, 'screeningQuestions'] = question_list
    
    #salary
    salary_dict = row['salary']
    salary_max = salary_dict['maximum']
    salary_min = salary_dict['minimum']
    salary_type = salary_dict['type']['salaryType']
    salary_maxs.append(salary_max)
    salary_mins.append(salary_min)
    salary_types.append(salary_type)

df_clean['company_uen'] = company_uens 
df_clean['company_name'] = company_names 
df_clean['company_description'] = company_descriptions 
df_clean['salary_max'] = salary_maxs 
df_clean['salary_min'] = salary_mins 
df_clean['salary_type'] = salary_types 
```
